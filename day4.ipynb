```python
with open("input/day4.txt", 'r') as f:
    input = [i.strip() for i in f.readlines()]

input
```

```python
total = 0

for entry in input:
    first, second = entry.split(',')
    first_start = int(first.split('-')[0])
    first_end = int(first.split('-')[1])

    second_start = int(second.split('-')[0])
    second_end = int(second.split('-')[1])

    if first_start <= second_start and first_end >= second_end:
        total += 1
    elif first_start >= second_start and first_end <= second_end:
        total += 1

total    
```

```python
total = 0

for entry in input:
    first, second = entry.split(',')
    first_start = int(first.split('-')[0])
    first_end = int(first.split('-')[1])

    second_start = int(second.split('-')[0])
    second_end = int(second.split('-')[1])

    if first_start < second_start and first_end < second_start:
        continue
    elif first_start > second_end and first_end > second_end:
        continue
    elif second_start < first_start and second_end < first_start:
        continue
    elif second_start > first_end and second_end > first_end:
        continue
    else:
        total += 1

total    
```
